## EXERCÍCIO 2

Faça um algoritmo que receba dois números e ao final mostre a soma, subtração, multiplicação e a divisão dos números lidos.

```python
n1 = int(input('Digite um numero:'))
n2 = int(input('Digite outro numero:'))

soma = n1 + n2
sub = n1 - n2
mult = n1 * n2

if n2 != 0:
    div = n1 / n2
    print(f'A soma, subtração, multiplicação e a divisão entre {n1} e {n2} são iguais a {soma}, {sub}, {mult} e {div}')
else:
    print(f'A soma, subtração e multiplicação entre {n1} e {n2} são iguais a {soma}, {sub} e {mult}')
    print('Divisão por zero não é permitida.')
```

## EXERCÍCIO 17

A concessionária de veículos “CARANGO VELHO” está vendendo os seus veículos com desconto. Faça um algoritmo que calcule e exiba o valor do desconto e o valor a ser pago pelo cliente de vários carros. O desconto deverá ser calculado de acordo com o ano do veículo (Até 2000 - 12% e acima de 2000 - 7%).

```python
resultado = []
for i in range(2):
    carro = int(input())
    ano = int(input())

    if ano <= 2000:
        soma = carro - (carro * 0.12)
    else:
        soma = carro - (carro * 0.07)

    resultado.append((soma, carro))

print(resultado)
```
