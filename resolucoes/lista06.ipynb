## Exercicio 1

Faça um programa que leia uma matriz 3 x 3. Imprima na tela o menor valor contido nessa matriz.

```python
matriz = []

for i in range(3):
    linha = []

    for j in range(3):
        linha.append(int(input(f'adicione o valor [{i}][{j}]: ')))
    
    matriz.append(linha)
    
menor = 1000

for linha in matriz:
    for celula in linha:
        if celula < menor:
            menor = celula

print(menor)
```
